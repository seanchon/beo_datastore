```python
# 288 Format for Load Data
# Energy Meter Export Channel

from load.customer.models import Meter

meter = Meter.objects.get(id=1)

# meter.export_channel.intervalframe.total_frame288.dataframe
# meter.import_channel.intervalframe.total_frame288.dataframe
# meter's interval frame is the sum of export and import channels
meter.intervalframe.total_frame288.dataframe
```

```python
# Clean Net Short - GHG Cost (tCO2/kWh)

from cost.ghg.models import GHGRate

cns = GHGRate.objects.get(name="Clean Net Short", effective__year=2018)
cns.frame288.dataframe
```

```python
# Example GHG Cost Calculation
# (288 Net Meter x 288 GHG CNS)

ghg_totals = meter.intervalframe.total_frame288.dataframe * GHGRate.objects.get(name="Clean Net Short", effective__year=2018).frame288.dataframe
ghg_totals
```

```python
# GHG Content from Building's Electricity Usage (tCO2/year)

ghg_totals = meter.intervalframe.total_frame288.dataframe * GHGRate.objects.get(name="Clean Net Short", effective__year=2018).frame288.dataframe
ghg_totals.sum().sum()
```
